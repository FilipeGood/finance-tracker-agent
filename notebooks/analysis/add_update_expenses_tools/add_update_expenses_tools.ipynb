```python
from langchain.agents import (
    create_tool_calling_agent,
    AgentExecutor,
    create_react_agent,
)
from langchain_openai import ChatOpenAI
from src.tools import (
    save_expense_tool,
    get_current_date_tool,
    get_all_expenses,
    update_last_expense_attribute,
)
from langchain_core.prompts import ChatPromptTemplate
from langchain import hub
from src.prompts import openai_functions_agent_prompt
from langchain.agents import AgentExecutor, create_openai_functions_agent


llm = ChatOpenAI(model="gpt-3.5-turbo")
```

```python
test_cases = [
    "Can you save a $10 expense for footbal game?",
    "Bought cigarrates for 10.15$",
    "Save a $50 gym membership expense",
    "Save a $25 dinner at restaurant",
    "Save a $80 grocery shopping expense",
]
```

```python
# Create agent with all tools including the correction tool
prompt = openai_functions_agent_prompt

# Make sure we have all tools: save, get_date, and update
tools = [
    save_expense_tool,
    get_current_date_tool,
    get_all_expenses,
    update_last_expense_attribute,
]

agent = create_openai_functions_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
```

```python
# Test with a regular expense save
result = agent_executor.invoke({"input": test_cases[4]})
print("Regular save test result:")
print(result["output"])
```
